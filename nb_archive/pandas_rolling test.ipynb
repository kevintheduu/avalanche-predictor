## I have never used pandas.rolling, let's practice and see what happens when rows are dropped

```python
import pandas as pd
```

```python
df = pd.read_csv('csv_files/stevens_pass_barometric_pressure_2014_now.csv')
```

```python
df.columns = [['Date','berne','haus']]
df.head()
```

```python
df['count'] = df['haus'].rolling(3).sum()
```

```python
df.head()
```

```python
df['d'] = df['count'].copy()
```

```python
df.head(10)
```

```python
df.drop(index=[0,1],axis=0)
```

## Conclusion
- Once called on, Pandas rolling takes a snapshot of the columns used for the numbers. It is not dynamic like excel functions that change when rows are deleted. This is great because now I know when to use it: before dropping rows!
